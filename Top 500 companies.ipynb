TOP 500 COMPANIES FINANCIAL ANALYSIS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
import seaborn as sns
```

```python
# LOAD THE DATA
df= pd.read_csv('/content/Financial Analytics data (1).csv')
```

```python
df.head(4)
```

```python
df.info()
```

```python
df.rename(columns={ 'Mar Cap - Crore':'Market_capitalization_crore', 'Sales Qtr - Crore':'Sales_qtr_crore'}, inplace=True)
```

```python
df.dropna(subset=['Market_capitalization_crore','Sales_qtr_crore'], inplace=True)
```

```python
df.info(4)
```

```python
# Derived metric
df['Market_cap_to_sales_ratio']= df.apply(lambda row: row['Market_capitalization_crore']/row['Sales_qtr_crore'] if row ['Sales_qtr_crore'] !=0 else np.nan, axis=1)
```

```python
#Key metric
print("\n Descriptive statistics metric for Market_capitalization: ")
print(df['Market_capitalization_crore'].describe())

print("\n Describe statistics metric for Sales_qtr:")
print(df['Sales_qtr_crore'].describe())

print("\n Describe statistics metric for Market_cap_to_sales_ratio:")
print(df['Market_cap_to_sales_ratio'].describe())
```

```python
# TOP 10 AND BOTTOM 10 COMPANIES BY MARKET CAPITALIZATION
print('\n Top 10 comanies by Market_capitalization')
Top_10_Market_capitalization = df.nlargest(10, "Market_capitalization_crore")[['Name', 'Market_capitalization_crore', 'Market_cap_to_sales_ratio']]
print(Top_10_Market_capitalization)
print('\n Botton 10 companies by Market_capitalization')
Bottom_10_Market_capitalization=df.nsmallest(10, 'Market_capitalization_crore')[['Name', 'Market_capitalization_crore', 'Market_cap_to_sales_ratio']]
print(Top_10_Market_capitalization)
```

```python
# TOP 10 AND BOTTOM 10 COMPANIES BY SALES_QTR
print('\n Top 10 comanies by Sales_Qtr')
Top_10_Sales_qtr = df.nlargest(10, "Sales_qtr_crore")[['Name', 'Sales_qtr_crore', 'Market_cap_to_sales_ratio']]
print(Top_10_Sales_qtr)
print('\n Botton 10 companies by Sales_Qtr')
Bottom_10_Sales_qtr=df.nsmallest(10, 'Sales_qtr_crore')[['Name', 'Sales_qtr_crore', 'Market_cap_to_sales_ratio']]
print(Bottom_10_Sales_qtr)
```

```python
# TOP 10 AND BOTTOM 1O MARKET_CAP_TO SALES_QTR_RATIO
print('\n Top 10 comanies by Market_cap_to_sales_ratio')
Top_10_Market_cap_to_sales_ratio = df.nlargest(10, "Market_cap_to_sales_ratio")[['Name', 'Market_cap_to_sales_ratio']]
print(Top_10_Market_cap_to_sales_ratio)
print('\n Botton 10 companies by Market_cap_to_sales_ratio')
Bottom_10_Market_cap_to_sales_ratio=df.nsmallest(10, 'Market_cap_to_sales_ratio')[['Name', 'Market_cap_to_sales_ratio']]
print(Bottom_10_Market_cap_to_sales_ratio)
```

```python
# co-relation matrix
correlation_matrix= df[['Market_capitalization_crore',"Sales_qtr_crore", "Market_cap_to_sales_ratio"]].corr()
print(correlation_matrix)
```

```python
# Visualization between Market_capitalization_crore and Sales_qtr_crore
plt.figure(figsize=(10,6))
plt.scatter(x='Market_capitalization_crore', y='Sales_qtr_crore', data=df, alpha=1.0, color= "yellow")
plt.title('Market_capitalization_crore vs Sales_qtr_crore')
plt.xlabel('Market_capitalization_crore')
plt.ylabel('Sales_qtr_crore')


plt.show()
```

```python
# Visualization of Market_capitalization and Market_cap_to_sales_ratio
plt.figure(figsize=(10,6))
plt.scatter(x='Market_capitalization_crore', y='Market_cap_to_sales_ratio', data=df, alpha=1.0, color='orange')
plt.title('Market_capitalization_crore vs Market_cap_to_sales_ratio')
plt.xlabel('Market_capitalization_crore')
plt.ylabel('Market_cap_to_sales_ratio')
plt.show()
```

```python
# Visualization of Sales_qtr and Market_cap_to_sales_ratio
plt.figure(figsize=(10,6))
plt.scatter(x='Sales_qtr_crore', y='Market_cap_to_sales_ratio', data=df, alpha=1.0, color='green')
plt.title('Sales_qtr_crore vs Market_cap_to_sales_ratio')
plt.xlabel('Sales_qtr_crore')
plt.ylabel('Market_cap_to_sales_ratio')
plt.show()
```

```python
# visualization of Market_cap_to_sales_ratio
plt.figure(figsize=(10,6))
plt.hist(df['Market_cap_to_sales_ratio'], bins=20, color='blue', edgecolor='black')
plt.title('Distribution of Market_cap_to_sales_ratio')
plt.xlabel('Market_cap_to_sales_ratio')
plt.ylabel('Frequency')
plt.show()
```
